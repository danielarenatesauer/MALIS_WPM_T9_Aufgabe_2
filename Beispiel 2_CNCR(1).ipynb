# 2. Beispiel: CNCR

Die chilenische Zeitschrift CNCR Conservación, Restauración y Patrimonio stellt auf ihrer Website eine Liste von links zu den bereits erschienen Nummern bereit:

```python
#es wird die Python Bibliothek beautiful soup importiert https://www.crummy.com/software/BeautifulSoup/bs4/doc/
#damit wird angestrebt, den Datenbaum zu strukturieren
from bs4 import BeautifulSoup
import requests
```

Daten werden von der Website der Zeitschrift CNCR importiert.

```python

URL="https://www.cncr.gob.cl/sitio/Secciones/Revista-Conserva/Numeros-publicados/"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify()) 
```

```python
# es werden alle urls herausgefiltert
for link in soup.find_all('a'):
    print("The text:", link.text)
    print("The URL:", link['href'])
    print("--------------")
```

```python
prefix = ("https://www.cncr.gob.cl/")
```

```python
for link in soup.find_all('a'):
    if "Revista-Conserva-N" in link['href']:
        print(prefix + link['href'])
```
